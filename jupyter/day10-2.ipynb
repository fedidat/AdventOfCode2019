```python
filename = 'inputs/day10.txt'
# filename = 'inputs/day10-test.txt'
# filename = 'inputs/day10-test2.txt'
with open(filename) as file:
    input = [list(l) for l in file.read().splitlines()]
```

```python
from collections import defaultdict
from math import gcd

def simplify(a, b):
    denominator = gcd(a,b)
    return (a/denominator, b/denominator)

def part1():
    ast = []
    for (i,l) in enumerate(input):
        for (j,a) in enumerate(l):
            if a == '#':
                ast.append((i,j))
    pairs = defaultdict(list)
    for a in ast:
        for b in [x for x in ast if x != a]:
            direction = simplify(b[1]-a[1], b[0]-a[0])
            if direction not in pairs[a]:
                pairs[a].append(direction)
    print("max", max(len(v) for v in pairs.values()))

part1()
```

```python
from math import atan2, pi
from collections import defaultdict

def part2():
    ast = []
    for (i,l) in enumerate(input):
        for (j,a) in enumerate(l):
            if a == '#':
                ast.append((j,i))
    pairs = defaultdict(list)
    for a in ast:
        for b in [x for x in ast if x != a]:
            pairs[a].append((b, (atan2(a[1]-b[1],a[0]-b[0])-pi/2)%(2*pi), (b[1]-a[1])**2+(b[0]-a[0])**2))
    print("max", max(len(set([direction for (c, direction, dist) in v])) for v in pairs.values()))
    
    station = (max(pairs, key=(lambda key: len(set([direction for (c, direction, dist) in pairs[key]])))))
#     print("station is at",station)
    rotation = sorted(pairs[station], key=lambda x: (x[1], x[2]))
#     print(rotation)
    i=0
    prev_direction = None
    while rotation:
        for (x, direction, dist) in rotation[:]:
            if prev_direction is None or prev_direction != direction:
                rotation.remove((x, direction, dist))
                prev_direction = direction
                i+=1
                if i == 200:
                    print("200th",x[0]*100+x[1])
                    return
        prev_direction = None

part2()
```
